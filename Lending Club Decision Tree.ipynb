Now i will start preparing the data to receive the decision tree model

```python
loans['purpose'].value_counts()
```

```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(loans, columns=cat_feats, drop_first=True)
```

```python
final_data.info()
```

Now is the time to separate the data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

After that it's time to add the DecisionTreeClassifier model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
pred = dtree.predict(X_test)
```

Let's check the accuracy of the model with classification_report and confusion_matrix

```python
from  sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pred))
```

This is already a relatively good model, but it still fails when it checks the data of people who would not pay it, it is also due to the fact that we have a lot of data from people who paid and few of people who did not pay, but it can get even better with trees random.

```python
print(confusion_matrix(y_test, pred))
```

With the aim of further improving the accuracy of the model, let's go to the random trees.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=500)
```

```python
rfc.fit(X_train, y_train)
```

```python
pred = rfc.predict(X_test)
```

Finally let's check again the accuracy of the new model.

```python
print(classification_report(y_test, pred))
```

The final model has its good side and its bad side, on the one hand it increased the level of prediction in some factors but on the other hand it sinned in data such as recall f1-score, so in this case the decision tree is even more viable.

```python
print(confusion_matrix(y_test, pred))
```
